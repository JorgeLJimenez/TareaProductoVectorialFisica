## DefiniciÃ³n de objetos

```python
import math #importo la librería math

class Vector: #defino la clase Vector

  def __init__(self, coordenada_x, coordenada_y):
    self.x = coordenada_x #defino las variables de clase
    self.y = coordenada_y
    self.magnitud = (self.x**2 + self.y**2)**0.5
    self.angulo = math.atan2(self.y, self.x)

  def __str__(self):
    respuesta = "( " + str(self.x) + " , " + str(self.y) + " )"
    return(respuesta)
    #este __srt__ viene de cadena, y es lo que quiero que saque al imprimir
    #sirve para que al print un vector, salga un vector y no algo raro
    #los dos def primeros son metodos especiales, para hacer cosas interesantes

  def suma(self, otra): #defino la función o método suma
    respuesta = Vector(self.x + otra.x, self.y + otra.y) #otra.x y otra.y???
    return(respuesta)

  def escalar(self, otra):
    return(self.x*otra.x + self.y*otra.y) #???
```

## TAREA. Aplicación del producto vectorial en la Física: Momento angular de una partícula clásica

```python
class Particula:
  #atributos
  def __init__(self, componente_x, componente_y, componente_z, velocidad_x, velocidad_y, velocidad_z, masa):
    self.x = componente_x
    self.y = componente_y
    self.z = componente_z
    self.v_x = velocidad_x
    self.v_y = velocidad_y
    self.v_z = velocidad_z
    self.m = masa

  #métodos
  def __str__(self):
    posicion = "[" + str(self.x) + "i + " + str(self.y) + "j + " + str(self.z) + "k] m"
    velocidad = "[" + str(self.v_x) + "i + " + str(self.v_y) + "j + " + str(self.v_z) + "k] m/s"
    masa = str(self.m) + " kg"
    informacion = "posición: " + posicion + "\n" + "velocidad: " + velocidad+ "\n" + "masa: " + masa
    return (informacion)
  
  def momento_angular(self, particula): #aqui se usa el producto vectorial
    L_x = particula.m*(particula.y*particula.v_z - particula.z*particula.v_y)
    L_y = particula.m*(particula.z*particula.v_x - particula.x*particula.v_z)
    L_z = particula.m*(particula.x*particula.v_y - particula.y*particula.v_z)

    resultado = "[" + str(L_x) + "i + " + str(L_y) + "j + " + str(L_z) + "k] (kg*m^2)/s"

    return(resultado)
```

```python
particula_1 = Particula(2,3,4,3,5,4,1)
particula_2 = Particula(4,2,1,7,1,0,2.5)
print("Partícula 1: \n", particula_1)
print("Momento angular de la partícula 1: ", particula_1.momento_angular(particula_1),"\n")
print("Partícula 2: \n", particula_2)
print("Momento angular de la partícula 2: ", particula_2.momento_angular(particula_2))
```
